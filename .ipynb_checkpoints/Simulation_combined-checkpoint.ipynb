# Name and Gender variable

```python
first_names_gender = (pd.read_csv('data/name_gender_dataset.csv'))[['Name', 'Gender']]
first_names_gender = first_names_gender.sample(n = 250, random_state=1)
first_names_gender = first_names_gender.rename(columns = {'Name' : 'First_Name'})
first_names_gender
```

```python
last_names = (pd.read_csv('data/most-common-name_surnames.csv'))['name']
last_names = pd.DataFrame(last_names)
last_names = last_names.sample(250, random_state=1)
last_names.columns = ['Last_Name']
last_names
```

# BMI and Height

```python
weight_height = pd.read_csv('data/weight-height.csv')
weight_height['Height'] = weight_height['Height']/100
weight_height['bmi'] = (weight_height['Weight']/(weight_height['Height'])**2)
weight_height = weight_height.drop('Gender', 1)
weight_height = weight_height.sample(250, random_state = 1)
weight_height
```

# Case-Control Status

```python
random.seed(1)
case_control = ['case', 'control']
case_control_df = random.choices(case_control, k = 250)
case_control_df = pd.DataFrame(case_control_df, columns = ['case_control status'])
case_control_df
```

# Education Level

```python
random.seed(1)
education_level = ['primary', 'high school', 'bachelor', 'master', 'phD']
education_df = random.choices(education_level, k = 250)
education_df = pd.DataFrame(education_df, columns = ['education level'])
education_df
```
